### Cleaning the Imported Data 

As, there are some missing values present in the data which could affect the decision making process thus, it is essential to clean the data. This is generally done using the following :

1. Drop - It is used to drop columns containing null values that exist in a Dataset.
2. Isnull - Used to check the total number of null values existing in the columns present in the dataset.
3. Dropna - It is used to drop any null values that exist in a Dataset

```python
df.drop(columns = ['Status', 'unnamed1'], inplace = True)
```

```python
df.info()
```

# Exploratory Data Analysis

### Classification on the basis of Gender & It's Count

```python
#plotting bar chart using Seaborn

ax = sns.countplot(x = 'Gender', data = df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue='Gender')
```

## Observation 1
*From the above 2 graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Classification on the basis of AGE

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue = 'Age Group')
```

## Observation 2

*It can be observed from above graphs that most of the buyers are of age group between 26-35 yrs female*

### Classification on the Basis of State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,4)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders',hue ='State')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,4)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount', hue='State')
```

## Observation 3

*From above graphs we see that most of the orders & (Total Sales/Amount) are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Classification on the basis of Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Marital_Status')

sns.set(rc={'figure.figsize':(2,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

## Observation 4

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Classification on the basis of Occupation


```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation', hue = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount', hue = 'Occupation')
```

## Observation 5

*From above graphs we see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Classification based on Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category', hue = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount', hue = 'Product_Category')
```

### Observation 6

*From above graphs we see that most of the sold products are from Food, Clothing and Electronics category*

### Classification on the basis of Products

```python
#Identifying the top 10 most sold product
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue = 'Product_ID')
```
